```python
import pandas as pd

df = pd.read_csv("./y.csv")
df = df.rename({"backbone": "MPAA", "pyridone": "Pyridone", "beta_av": "Yield"}, axis=1)
df.loc[df["MPAA"] == 19, "MPAA"] = 6
df.loc[df["MPAA"] == 20, "MPAA"] = 7
df
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.dpi"] = 160
plt.rcParams["font.size"] = 8
```

```python
df = df.sort_values(["Pyridone", "MPAA"])
entries = df["Pyridone"].unique()
dfs = [df.loc[df["Pyridone"] == entry, "Yield"] for entry in entries]
colors = [plt.cm.jet(i/len(entries)) for i in range(len(entries))]
labels = [f"Pyridone{entry}"for entry in entries]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 3))
ax.hist(dfs, bins=30, stacked=True, color=colors, label=labels)
ax.set_xticks(range(0, 101, 10))
ax.set_xlim(0, 100)
ax.set_xlabel("Yield (%)")
ax.set_ylabel("Count")
plt.legend()
plt.savefig("yield_pyridone.png", bbox_inches="tight")
plt.show()
```
